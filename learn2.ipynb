## Python Natural Language Processing Toolkit


```python
import nltk
from nltk.corpus import gutenberg 
from nltk.text import Text 
from nltk.probability import FreqDist
import random
```

```python
nltk.download('gutenberg')
```

```python
#NLTK text Processing 
text  = Text(gutenberg.words('shakespeare-hamlet.txt'))
```

```python
#Corpus vs Corpora
corpus = nltk.corpus.gutenberg 
corpora_list = corpora_list = corpus.fileids()

print(corpus)
print(corpora_list)
```

```python
#Accessing Text Corpora with NLTK
sample_text = corpus.raw('shakespeare-hamlet.txt')
print(sample_text)
```

```python
#Corcordance, Text Similarity & Common Contexts
text.concordance("Hamlet")
text.similar("Hamlet")
text.common_contexts(["Hamlet","king"])
```

```python
#Plotting Word Distribution Graph
fdist = FreqDist(text)
fdist.plot(30)
```

```python
#Token vs Types , Lexicon vs Grammar , Tokenization
tokens = nltk.word_tokenize(sample_text)
types = set(tokens)
print("Token Count:",len(tokens))
print("Type Count:",len(types))
```

```python
#Zipf Law & Frequency Distribution
fdist.plot(30, cumulative=True)
```
